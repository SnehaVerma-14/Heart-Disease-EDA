```python
import pandas as pd

df = pd.read_csv("processed.cleveland.data", header=None)
df.head()
```

```python
df.columns = [
    "age",         # Age of the patient
    "sex",         # 1 = Male, 0 = Female
    "cp",          # Chest pain type
    "trestbps",    # Resting blood pressure
    "chol",        # Serum cholesterol
    "fbs",         # Fasting blood sugar > 120 mg/dl
    "restecg",     # Resting ECG results
    "thalach",     # Maximum heart rate achieved
    "exang",       # Exercise-induced angina
    "oldpeak",     # ST depression induced by exercise
    "slope",       # Slope of peak exercise ST segment
    "ca",          # Number of major vessels colored by fluoroscopy
    "thal",        # Thalassemia test result
    "target"       # Heart disease diagnosis
]

df.head()
```

Step 6: Analyze the Target Variable (Heart Disease)

```python
df["target"].value_counts()
```

```python
df["heart_disease"] = df["target"].apply(lambda x: 0 if x == 0 else 1)
```

```python
df["heart_disease"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["heart_disease"].value_counts().plot(kind="bar")
plt.title("Heart Disease Distribution")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Number of Patients")
plt.show()
```

Step 7: Age Analysis

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["age"], bins=10, edgecolor="black")
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x="heart_disease", y="age", data=df)
plt.title("Age vs Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
df.groupby("heart_disease")["age"].mean()
```

Step 8: Gender Analysis (Male vs Female)

```python
df["sex"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(x="sex", data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="heart_disease", data=df)
plt.title("Heart Disease by Gender")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.legend(title="Heart Disease")
plt.show()
```

```python
pd.crosstab(
    df["sex"],
    df["heart_disease"],
    normalize="index"
) * 100
```
